```python
def datetime_to_ntp(ts):
    """ Convert python datetime into Internet Network Time Protocol (NTP) 64bit timestamp """
    diff = ts - datetime.datetime(1900, 1, 1, 0, 0, 0, 0, pytz.utc)
    return np.uint64(((diff.days*24*60*60 + diff.seconds) << 32) + ts.microsecond // 2**32)
```
